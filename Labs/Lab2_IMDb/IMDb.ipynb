## Sobre el dataset

IMBD es una base de datos online de información relacionada a películas, series de televisión, videos, entre otros. El dataset contiene información de cerca de 5000 titulos y metadatos que van desde el director, actores hasta scores de críticos e impacto en redes sociales 

### Fuente

### Descripción de columnas

| Variable Name| Description |
| --- | --- | 
| movie_title | Title of the Movie | 
|duration |	Duration in minutes |
|director_name |	Name of the Director of the Movie |
|director_facebook_likes |	Number of likes of the Director on his Facebook Page
|actor_1_name |	Primary actor starring in the movie
|actor_1_facebook_likes |	Number of likes of the Actor_1 on his/her Facebook Page

|num_voted_users 	|Number of people who voted for the movie
|cast_total_facebook_likes |	Total number of facebook likes of the entire cast of the movie
|movie_facebook_likes |	Number of Facebook likes in the movie page
|plot_keywords |	Keywords describing the movie plot
|facenumber_in_poster |	Number of the actor who featured in the movie poster

# Determina la cantidad de películas que fueron producidas cada año, de 2006 a 2016

```python
dfCantPelis = dfCantPelis[(dfCantPelis.title_year > 2005) & (dfCantPelis.title_year < 2017)]
grouped_data = dfCantPelis.groupby('title_year').title_year.count()
print(grouped_data)

#Graficar
grouped_data.plot(
    x='title_year', 
    y='title_year', 
    kind='bar', 
    legend=False, 
    color='blue',
    width=0.8
)
#Style
plt.ylabel("Número de películas producidas")
plt.xlabel("Año")
plt.title("Películas producidas de 2006 a 2016")
plt.gca().yaxis.grid(linestyle=':')
```

# Análisis de Rentabilidad

La recaudación es importante, pero también lo es la rentabilidad, es decir que tantas ganacias me genera respecto a mi inversión.

```python
#Limpieza
dfRentabilidad = dfDocumentales.copy()

dfRentabilidad = dfRentabilidad[dfDocumentales.budget.notnull()]
dfRentabilidad['budget']=dfRentabilidad.budget.astype(int)
```

##  Presupuesto y Recaudación

```python
dfGrossBudget = dfRentabilidad[(dfRentabilidad['genres'].str.contains('Documentary')) & (dfRentabilidad.title_year > 2000)]

dfGrossBudget = dfGrossBudget[['director_name', 
                               'movie_title', 'gross','budget','title_year']]
dfGrossBudget.head(10)
```

# Top de Ganancias y Rentabilidad

```python
dfGrossBudget['profits'] = dfGrossBudget['gross'] - dfGrossBudget['budget']
dfGrossBudget["profitability"] = dfGrossBudget["gross"] / dfGrossBudget["budget"]
dfGrossBudget.sort_values('profitability', ascending=False).head(10)
```

### --->Mayor Rentabilidad: Jonathan Caouette 2715 por cada dólar invertido

Aquí tenemos un punto muy importante, Directores rentables como Jonathan Caouette que con 218 dólares recaudo casi 600 mil, es decir por cada dólar invertido ¡recaudó 2715!!! Spurlock recaudó 177 por cada dólar. En 4ta posición está Moore que en 2004 recaudó casi 20 dólares por cada dólar invertdio. Tremaine está en 7ma posición con 12.85 dólares.
